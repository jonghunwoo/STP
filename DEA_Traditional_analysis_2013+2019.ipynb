```
class DEA():

    def __init__(self,csv_name):
        df_original = pd.read_csv(csv_name)
        self.df = df_original[['Shipyard','신조선건조량 ','신조선수주량','직원수','안벽길이(m)','Dock 및 선대 개수 (기)']]
        self.df.dropna(axis=0,how='any',inplace=True)
        self.df_list = (self.df).values
        self.dmu_value = []

    def DMU_caculate(self):
        j = 1
        z = self.df
        x = self.df_list
        v1 = cp.Variable()
        v2 = cp.Variable()
        v3 = cp.Variable()
        u1 = cp.Variable()
        u2 = cp.Variable()
        for i in range (len(z.index)):
            constraints = [x[i, j] * u1 + x[i, j+1] * u2 == 1,
                           u1 >= 0,
                           u2 >= 0,
                           v1 >= 0,
                           v2 >= 0,
                           v3 >= 0]
            for k in range(len(z.index)):
                constraints.append(x[k, j] * u1 + x[k, j+1] * u2 - x[k, j+2] * v1 - x[k, j+3] * v2 - x[k, j+4] * v3 <= 0)

            m = x[i, j + 2] * v1 + x[i, j + 3] * v2 + x[i, j + 4] * v3
            obj = cp.Minimize(m)
            prob = cp.Problem(obj, constraints)
            prob.solve()
            self.dmu_value.append(m.value)
        return self.dmu_value

    def DMU_Efficiency_caculate(self):
        e = []
        for s in range(len(self.dmu_value)):
            f = 1 / self.dmu_value[s]
            e.append(f)
        return e


n = DEA("/content/gdrive/My Drive/연구실 과제/DEA/작업/" +"DEA_Traditional analysis_2013.csv" )

print(n.DMU_caculate())

print(n.DMU_Efficiency_caculate())
```
